```python
# Function to get categorical and numerical columns
def split_num_cat_cols(df):
    cat_cols= []
    num_cols= []

    for column_name in df:
        if df[column_name].dtype == 'int64' or df[column_name].dtype == 'float64':
            num_cols.append(column_name)
        else:
            cat_cols.append(column_name)
            
    return num_cols,cat_cols
```

```python
import pandas as pd
pd.options.display.max_columns = 50

df = pd.read_feather('datasets/before_enccoding_scaling.feather')
df.head()
```

```python
# Handling date features
# These dates are only month and year so we are just dividinng them

df['issue_d'] = df['issue_d'].ffill()
df['issue_d'] = '01-'+df['issue_d']
df['issue_d'] = df['issue_d'].apply(pd.to_datetime)
df['issue_d_month'] = df['issue_d'].dt.month
df['issue_d_year'] = df['issue_d'].dt.year

df.drop('issue_d', axis=1, inplace=True)
df.issue_d_month = df.issue_d_month.astype('int64')
df.issue_d_year = df.issue_d_year.astype('object')

df.head()
```

```python
# Encoding of categorical features
# 2 types of encoding is used here:       1. Binary encoding       2. Ordinal Encoding
num_cols,cat_cols = split_num_cat_cols(df)


# 1. Binary Encoding
import category_encoders as ce

bin_encod_emp_title = ce.BinaryEncoder(cols=['emp_title'])
df_emp_title = bin_encod_emp_title.fit_transform(df['emp_title'])
df = pd.concat([df, df_emp_title], axis=1)

bin_encod_purpose = ce.BinaryEncoder(cols=['purpose'])
df_purpose = bin_encod_purpose.fit_transform(df['purpose'])
df = pd.concat([df, df_purpose], axis=1)

bin_encod_addr_state = ce.BinaryEncoder(cols=['addr_state'])
df_addr_state = bin_encod_addr_state.fit_transform(df['addr_state'])
df = pd.concat([df, df_addr_state], axis=1)

df.drop(['emp_title','purpose','addr_state'], axis=1, inplace=True)


# 2. Ordinal Encoding
cols_dict = {}
cols_dict['term'] = {' 36 months': 1, ' 60 months': 2}
cols_dict['emp_length'] = {'10+ years':10, '6 years':6, '4 years':4, '< 1 year':0.5, '2 years':2,'9 years':9, '5 years':5, '3 years':3, '7 years':7, '1 year':1,'8 years':8}
cols_dict['sub_grade'] = {'G5': 0, 'G4': 1, 'G3': 2, 'G2': 3, 'G1': 4,
                    'F5': 5, 'F4': 6, 'F3': 7, 'F2': 8, 'F1': 9,
                    'E5': 10, 'E4': 11, 'E3': 12, 'E2': 13, 'E1': 14,
                    'D5': 15, 'D4': 16, 'D3': 17, 'D2': 18, 'D1': 19,
                    'C5': 20, 'C4': 21, 'C3': 22, 'C2': 23, 'C1': 24,
                    'B5': 25, 'B4': 26, 'B3': 27, 'B2': 28, 'B1': 29,
                    'A5': 30, 'A4': 31, 'A3': 32, 'A2': 33, 'A1': 34}    
cols_dict['home_ownership'] = {'MORTGAGE': 4, 'RENT': 3, 'OWN': 5,'ANY': 2,'OTHER': 1, 'NONE':0 }
cols_dict['verification_status'] = {'Source Verified':2,'Verified': 1,'Not Verified': 0}
cols_dict['pymnt_plan'] = {'n':0, 'y':1}
cols_dict['application_type'] = {'Individual':0, 'Joint App':1}
cols_dict['loan_status'] = {'Fully Paid':1, 'Current':0, 'Charged Off':0, 'In Grace Period':0,'Late (31-120 days)':0, 'Late (16-30 days)':0, 'Default':0,'Does not meet the credit policy. Status:Fully Paid':1,'Does not meet the credit policy. Status:Charged Off':0}
cols_dict['disbursement_method'] = {'Cash':0, 'DirectPay':1}
cols_dict['issue_d_year'] = {2015:9, 2018:12, 2017:11, 2016:10, 2014:8, 2011:5, 2010:4, 2009:3, 2008:2, 2007:1, 2013:7, 
       2012:6, 2019:13, 2020:14, 2021:15}

for i in set(cat_cols)-set(['emp_title','purpose','title','addr_state','issue_d_month']):
    df[i] = df[i].map(cols_dict[i])
    


# Storing all the models to predict further with these same parameters
import pickle

with open('models/encoding/dict_encoding.pickle', 'wb') as handle:
    pickle.dump(cols_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)

pickle_out = open('models/encoding/bin_encod_emp_title.pkl','wb')
pickle.dump(bin_encod_emp_title, pickle_out)
pickle_out = open('models/encoding/bin_encod_purpose.pkl','wb')
pickle.dump(bin_encod_purpose, pickle_out)
pickle_out = open('models/encoding/bin_encod_addr_state.pkl','wb')
pickle.dump(bin_encod_addr_state, pickle_out)
```

```python
def fit_score_clf(clf, X_train, y_train, X_test, y_test, name='clf'):
    """
    fits a classifer from sklearn and returns a dataframe
    clf(sklearn): classifier
    X_train(numpy): train features
    y_train(numpy): predictions row vector
    X_test(numpy): test features
    y_test(numpy): ground truth row vector
    name(string): name of classifier
    
    """
    #fit to train data
    clf.fit(X_train, y_train)
    #make predictions
    y_pred = clf.predict(X_test)
    #make dataframe from report
    clf_report = pd.DataFrame(classification_report(y_test, y_pred, output_dict=True))
    #add clf name as column
    clf_report['clf'] = name
    
    return clf_report
```

```python
import pandas as pd
pd.options.display.max_columns = 72
pd.set_option('display.max_rows', 50)

df = pd.read_feather('datasets/after_enccoding_scaling.feather')
df.loan_status = df.loan_status.astype('int64')
df.head()
```

```python
X = df.drop('loan_status', axis=1)
y = df.loan_status


from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=21)
```

```python
#instantiate with default params
from sklearn.linear_model import LogisticRegression

lr  = LogisticRegression(penalty='l2')
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)


import pickle

pickle_out = open('models/models/model_linear_regression.pkl','wb')
pickle.dump(lr, pickle_out)


from sklearn.metrics import classification_report

LogisticRegression_report = pd.DataFrame(classification_report(y_test, y_pred, output_dict=True))
LogisticRegression_report
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gbtree = GradientBoostingClassifier(learning_rate=0.1, n_estimators = 20)
gbtree.fit(X_train,y_train)
y_pred = lr.predict(X_test)


import pickle

pickle_out = open('models/models/model_GradientBoostingClassifier.pkl','wb')
pickle.dump(gbtree, pickle_out)


from sklearn.metrics import classification_report

GradientBoostingClassifier_report = pd.DataFrame(classification_report(y_test, y_pred, output_dict=True))
GradientBoostingClassifier_report
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=20, n_jobs=-1)
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)


import pickle

pickle_out = open('models/models/model_RandomForestClassifier.pkl','wb')
pickle.dump(rf, pickle_out)


from sklearn.metrics import classification_report

RandomForestClassifier_report = pd.DataFrame(classification_report(y_test, y_pred, output_dict=True))
RandomForestClassifier_report
```

```python
import xgboost as xgb

xgb_clf = xgb.XGBClassifier(n_estimators=20, subsample=0.4, n_jobs=-1, use_label_encoder=False)
xgb_clf.fit(X_train, y_train)
y_pred = xgb_clf.predict(X_test)


import pickle

pickle_out = open('models/models/model_xgb.pkl','wb')
pickle.dump(xgb_clf, pickle_out)


from sklearn.metrics import classification_report

xgb_report = pd.DataFrame(classification_report(y_test, y_pred, output_dict=True))
xgb_report
```

```python
from sklearn.svm import SVC

svc = SVC(C=1.0, kernel='rbf', gamma='auto', probability=True, max_iter=200)
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)


import pickle

pickle_out = open('models/models/model_SVC.pkl','wb')
pickle.dump(svc, pickle_out)



from sklearn.metrics import classification_report

SVC_report = pd.DataFrame(classification_report(y_test, y_pred, output_dict=True))
SVC_report
```
